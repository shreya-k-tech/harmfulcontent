```python
# Cell 3 - helper functions for feature extraction (frames + mel)
def safe_listdir(path):
    try:
        return os.listdir(path)
    except FileNotFoundError:
        return []

def extract_frames_equal_spacing(video_path, num_frames=FRAMES_PER_CLIP, target_size=FRAME_SIZE):
    cap = cv2.VideoCapture(str(video_path))
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total <= 0:
        cap.release()
        # return a zero array shaped (num_frames, H, W, 3)
        return np.zeros((num_frames, target_size[1], target_size[0], 3), dtype=np.float32)
    idxs = np.linspace(0, max(total-1,0), num_frames, dtype=int)
    frames = []
    for i in idxs:
        cap.set(cv2.CAP_PROP_POS_FRAMES, int(i))
        ret, frame = cap.read()
        if not ret or frame is None:
            frame = np.zeros((target_size[1], target_size[0], 3), dtype=np.uint8)
        else:
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame = cv2.resize(frame, target_size)
        frame = frame.astype(np.float32) / 255.0
        frames.append(frame)
    cap.release()
    return np.stack(frames)  # (F,H,W,3)

def extract_audio_mel(video_path, sr=AUDIO_SR, n_mels=N_MELS, time_frames=MEL_TIME_FRAMES):
    # Try to load audio from file using librosa (librosa uses ffmpeg when video provided)
    try:
        y, _ = librosa.load(str(video_path), sr=sr)
    except Exception as e:
        # fallback: return zeros if audio cannot be read (prints for debugging)
        print(f"⚠️ librosa failed to load audio for {video_path}: {e}")
        return np.zeros((n_mels, time_frames), dtype=np.float32)
    if y.size == 0:
        return np.zeros((n_mels, time_frames), dtype=np.float32)
    mel = librosa.feature.melspectrogram(y, sr=sr, n_mels=n_mels, hop_length=512, n_fft=1024)
    mel_db = librosa.power_to_db(mel, ref=np.max)
    # pad/trim time axis to time_frames
    if mel_db.shape[1] < time_frames:
        pad_width = time_frames - mel_db.shape[1]
        mel_db = np.pad(mel_db, ((0,0),(0,pad_width)), mode='constant', constant_values=(mel_db.min(),))
    else:
        mel_db = mel_db[:, :time_frames]
    return mel_db.astype(np.float32)
```

```python
# Cell 5 - Preprocessing: extract frames & mels for all videos (run once)
# WARNING: can take long depending on dataset size. You can interrupt and re-run later; already processed files are skipped.

def preprocess_all(base_dir=BASE_DIR, splits=SPLITS, classes=CLASSES):
    processed = 0
    for split in splits:
        for cls in classes:
            folder = os.path.join(base_dir, split, cls)
            if not os.path.isdir(folder):
                continue
            for fname in tqdm(safe_listdir(folder), desc=f"Preprocessing {split}/{cls}"):
                if not fname.lower().endswith(('.mp4','.mov','.avi','.mkv')):
                    continue
                src = os.path.join(folder, fname)
                stem = f"{split}_{cls}_{os.path.splitext(fname)[0]}"
                frames_out = os.path.join(FRAMES_DIR, stem + ".npy")
                mels_out   = os.path.join(MELS_DIR, stem + ".npy")
                if os.path.exists(frames_out) and os.path.exists(mels_out):
                    continue  # skip already done
                try:
                    frames = extract_frames_equal_spacing(src, num_frames=FRAMES_PER_CLIP, target_size=FRAME_SIZE)
                    mel = extract_audio_mel(src)
                    np.save(frames_out, frames)
                    np.save(mels_out, mel)
                    processed += 1
                except Exception as e:
                    print(f"Error processing {src}: {e}")
    print(f"Done. New processed files: {processed}")

# Run the preprocessing. If your dataset is large, this will take time.
preprocess_all()
```

```python
from torch.utils.data import Dataset, DataLoader
import numpy as np
import torch
import os

class HarmfulMultiModalDataset(Dataset):
    def __init__(self, frames_dir, mels_dir):
        self.frame_files = []
        self.mel_files = []
        self.labels = []
        self.label_names = []

        # List all frame npy files
        for file in os.listdir(frames_dir):
            if file.endswith(".npy"):
                frame_path = os.path.join(frames_dir, file)
                mel_path = os.path.join(mels_dir, file)  # same filename expected
                if os.path.exists(mel_path):
                    self.frame_files.append(frame_path)
                    self.mel_files.append(mel_path)

                    label = file.split("_")[1] if len(file.split("_")) > 1 else "Unknown"
                    self.labels.append(label)
                    if label not in self.label_names:
                        self.label_names.append(label)

        self.label_to_id = {lbl: i for i, lbl in enumerate(self.label_names)}

    def __len__(self):
        return len(self.frame_files)

    def __getitem__(self, idx):
        frames = np.load(self.frame_files[idx])
        mel = np.load(self.mel_files[idx])
        label = self.label_to_id[self.labels[idx]]

        frames = torch.tensor(frames, dtype=torch.float32).permute(3, 0, 1, 2)  # (C, T, H, W)
        mel = torch.tensor(mel, dtype=torch.float32)
        label = torch.tensor(label, dtype=torch.long)

        return {"frames": frames, "mel": mel, "label": label}

# ✅ Paths
frames_dir = r"E:\archive\Dataset\processed_data\frames"
mels_dir = r"E:\archive\Dataset\processed_data\mels"

dataset = HarmfulMultiModalDataset(frames_dir, mels_dir)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True, num_workers=0)

print(f"✅ Dataset Ready: {len(dataset)} samples, Classes: {dataset.label_names}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class VideoAudioNet(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        # 🎥 Video branch
        self.video_cnn = nn.Sequential(
            nn.Conv3d(3, 16, kernel_size=(3,3,3), padding=1),
            nn.ReLU(),
            nn.MaxPool3d((1,2,2)),
            nn.Conv3d(16, 32, kernel_size=(3,3,3), padding=1),
            nn.ReLU(),
            nn.MaxPool3d((2,2,2))
        )

        # 🔊 Audio branch
        self.audio_cnn = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        # ⚙️ Fusion layers
        self.fc = nn.Sequential(
            nn.Linear(32*7*7 + 32*16*28, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, frames, mel):
        x_v = self.video_cnn(frames)       # (B, 32, T/2, H/4, W/4)
        x_v = x_v.mean([2,3,4])            # global pooling

        x_a = self.audio_cnn(mel.unsqueeze(1))
        x_a = x_a.view(x_a.size(0), -1)

        x = torch.cat((x_v, x_a), dim=1)
        return self.fc(x)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class VideoAudioNet(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        # 🎥 Video branch
        self.video_cnn = nn.Sequential(
            nn.Conv3d(3, 16, kernel_size=(3,3,3), padding=1),
            nn.ReLU(),
            nn.MaxPool3d((1,2,2)),
            nn.Conv3d(16, 32, kernel_size=(3,3,3), padding=1),
            nn.ReLU(),
            nn.MaxPool3d((2,2,2))
        )

        # 🔊 Audio branch
        self.audio_cnn = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        # 🧮 Determine flattened size dynamically
        with torch.no_grad():
            dummy_frames = torch.zeros(1, 3, 16, 112, 112)   # (B, C, T, H, W)
            dummy_mel = torch.zeros(1, 1, 128, 224)          # (B, 1, MelBins, Time)
            v_feat = self.video_cnn(dummy_frames)
            a_feat = self.audio_cnn(dummy_mel)
            v_flat = v_feat.mean([2,3,4])                    # global pooling
            a_flat = a_feat.view(a_feat.size(0), -1)
            fusion_dim = v_flat.shape[1] + a_flat.shape[1]

        # ⚙️ Fusion layers
        self.fc = nn.Sequential(
            nn.Linear(fusion_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, frames, mel):
        x_v = self.video_cnn(frames)       # (B, 32, T/2, H/4, W/4)
        x_v = x_v.mean([2,3,4])            # global pooling (B, 32)

        x_a = self.audio_cnn(mel.unsqueeze(1))
        x_a = x_a.view(x_a.size(0), -1)

        x = torch.cat((x_v, x_a), dim=1)
        return self.fc(x)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score
from tqdm import tqdm
import numpy as np

# ==========================================
# 1️⃣ Dataset (simulated example)
# ==========================================
class VideoAudioDataset(Dataset):
    def __init__(self, num_samples=100, num_frames=16, height=112, width=112, num_classes=2):
        self.videos = np.random.randn(num_samples, num_frames, height, width, 3).astype(np.float32)
        self.labels = np.random.randint(0, num_classes, num_samples)
        self.label_names = ["Non-Harmful", "Harmful"]

    def __len__(self):
        return len(self.videos)

    def __getitem__(self, idx):
        frames = torch.tensor(self.videos[idx])          # (T, H, W, C)
        label = torch.tensor(self.labels[idx]).long()
        return frames, label


# ==========================================
# 2️⃣ Model
# ==========================================
class VideoAudioNet(nn.Module):
    def __init__(self, num_classes=2):
        super(VideoAudioNet, self).__init__()

        # Video feature extractor (Conv3D)
        self.video_cnn = nn.Sequential(
            nn.Conv3d(3, 16, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool3d((2, 2, 2)),

            nn.Conv3d(16, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool3d((2, 2, 2))
        )

        # Audio feature extractor (Conv2D)
        self.audio_cnn = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        # Fully connected layer for fusion
        self.fc = nn.Sequential(
            nn.Linear(32 + 32 * 32 * 56, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, frames, mel):
        # ✅ Fix shape: (B, T, H, W, C) → (B, C, T, H, W)
        if frames.ndim == 5 and frames.shape[-1] == 3:
            frames = frames.permute(0, 4, 1, 2, 3)

        # Video path
        x_v = self.video_cnn(frames)
        x_v = x_v.mean([2, 3, 4])  # Global average pooling → (B, 32)

        # Audio path
        x_a = self.audio_cnn(mel)
        x_a = x_a.view(x_a.size(0), -1)  # Flatten

        # Fusion
        x = torch.cat((x_v, x_a), dim=1)
        out = self.fc(x)
        return out


# ==========================================
# 3️⃣ Setup
# ==========================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using:", device)

dataset = VideoAudioDataset()
train_loader = DataLoader(dataset, batch_size=4, shuffle=True, num_workers=0)

num_classes = len(dataset.label_names)
model = VideoAudioNet(num_classes).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
epochs = 5

train_losses, accuracies = [], []

# ==========================================
# 4️⃣ Training Loop
# ==========================================
for epoch in range(epochs):
    model.train()
    epoch_loss, preds, trues = [], [], []

    for frames, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
        frames, labels = frames.to(device), labels.to(device)

        # Fake mel spectrogram for now (1×128×224)
        mel = torch.randn(frames.size(0), 1, 128, 224).to(device)

        optimizer.zero_grad()
        out = model(frames, mel)
        loss = F.cross_entropy(out, labels)
        loss.backward()
        optimizer.step()

        epoch_loss.append(loss.item())
        preds += out.argmax(1).cpu().tolist()
        trues += labels.cpu().tolist()

    acc = accuracy_score(trues, preds)
    train_losses.append(np.mean(epoch_loss))
    accuracies.append(acc)
    print(f"✅ Epoch {epoch+1}: Loss={np.mean(epoch_loss):.4f}, Acc={acc:.4f}")

# ==========================================
# 5️⃣ Save Model
# ==========================================
torch.save(model.state_dict(), "harmful_model.pt")
print("✅ Model saved as harmful_model.pt")
```

```python
# ============================================================
# STEP 1: Import Libraries
# ============================================================
import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import seaborn as sns
import random

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# ============================================================
# STEP 2: Dataset Loader
# ============================================================
class HarmfulDataset(Dataset):
    def __init__(self, frames_dir, mels_dir):
        self.samples = []
        self.frames_dir = frames_dir
        self.mels_dir = mels_dir
        self.categories = set()

        for fname in os.listdir(frames_dir):
            if not fname.endswith(".npy"):
                continue

            parts = fname.split("_")
            label_name = parts[1].strip() if len(parts) > 1 else "Unknown"

            frame_path = os.path.join(frames_dir, fname)
            mel_path = os.path.join(mels_dir, fname)
            if os.path.exists(mel_path):
                self.samples.append((frame_path, mel_path, label_name))
                self.categories.add(label_name)

        self.categories = sorted(list(self.categories))
        self.label_map = {c: i for i, c in enumerate(self.categories)}

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        frame_path, mel_path, label_name = self.samples[idx]
        frames = np.load(frame_path)
        mel = np.load(mel_path)

        frames = frames.astype(np.float32) / 255.0
        mel = mel.astype(np.float32)

        frames = torch.tensor(frames).permute(3, 0, 1, 2)  # (C, T, H, W)
        mel = torch.tensor(mel).unsqueeze(0)               # (1, H, W)
        label = torch.tensor(self.label_map[label_name]).long()

        return frames, mel, label

# ============================================================
# STEP 3: Define the Deep Learning Model
# ============================================================
class VideoAudioNet(nn.Module):
    def __init__(self, num_classes):
        super(VideoAudioNet, self).__init__()
        self.video_cnn = nn.Sequential(
            nn.Conv3d(3, 16, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool3d(2),
            nn.Conv3d(16, 32, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool3d(2),
        )

        self.audio_cnn = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(16, 32, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
        )

        self.fc = nn.Sequential(
            nn.Linear(32*2, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, frames, mel):
        x_v = self.video_cnn(frames)
        x_v = x_v.mean([2,3,4])  # Global average pool

        x_a = self.audio_cnn(mel)
        x_a = x_a.mean([2,3])

        x = torch.cat([x_v, x_a], dim=1)
        out = self.fc(x)
        return out

# ============================================================
# STEP 4: Load Dataset
# ============================================================
frames_dir = r"E:\archive\Dataset\processed_data\frames"
mels_dir   = r"E:\archive\Dataset\processed_data\mels"

dataset = HarmfulDataset(frames_dir, mels_dir)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_ds, val_ds = random_split(dataset, [train_size, val_size])

train_loader = DataLoader(train_ds, batch_size=4, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=4, shuffle=False)

print(f"✅ Dataset ready: {len(dataset)} samples across {len(dataset.categories)} classes -> {dataset.categories}")

# ============================================================
# STEP 5: Training Setup
# ============================================================
model = VideoAudioNet(num_classes=len(dataset.categories)).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
epochs = 5

# ============================================================
# STEP 6: Training Loop
# ============================================================
train_losses, val_losses, val_accs = [], [], []

for epoch in range(epochs):
    model.train()
    running_loss = 0
    for frames, mel, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
        frames, mel, labels = frames.to(device), mel.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(frames, mel)
        loss = F.cross_entropy(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    avg_train_loss = running_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # Validation
    model.eval()
    val_loss, correct, total = 0, 0, 0
    with torch.no_grad():
        for frames, mel, labels in val_loader:
            frames, mel, labels = frames.to(device), mel.to(device), labels.to(device)
            outputs = model(frames, mel)
            loss = F.cross_entropy(outputs, labels)
            val_loss += loss.item()
            preds = outputs.argmax(dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    avg_val_loss = val_loss / len(val_loader)
    acc = correct / total
    val_losses.append(avg_val_loss)
    val_accs.append(acc)

    print(f"Epoch {epoch+1}: Train Loss={avg_train_loss:.4f}, Val Loss={avg_val_loss:.4f}, Val Acc={acc*100:.2f}%")

# ============================================================
# STEP 7: Visualization
# ============================================================
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.legend()
plt.title("Loss Over Epochs")

plt.subplot(1,2,2)
plt.plot(val_accs, label='Validation Accuracy', color='green')
plt.legend()
plt.title("Validation Accuracy")
plt.show()

# ============================================================
# STEP 8: Evaluation
# ============================================================
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for frames, mel, labels in val_loader:
        frames, mel = frames.to(device), mel.to(device)
        outputs = model(frames, mel)
        preds = outputs.argmax(dim=1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(labels.numpy())

# Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dataset.categories)
disp.plot(cmap="Blues", xticks_rotation=45)
plt.title("Confusion Matrix")
plt.show()

# Classification Report
print("\nClassification Report:")
print(classification_report(all_labels, all_preds, target_names=dataset.categories))

# ============================================================
# STEP 9: Prevention System Logic (Example)
# ============================================================
def prevent_action(pred_label, confidence):
    if pred_label in ["Adult Content", "Harmful Content", "Suicide"] and confidence > 0.8:
        print(f"⚠️ ALERT: {pred_label} detected with {confidence*100:.1f}% confidence.")
        print("Action: Block video, show warning, and suggest reporting.")
    else:
        print("✅ Content is Safe to View.")

# Example of Prevention Step
sample_frame, sample_mel, _ = random.choice(dataset)
model.eval()
with torch.no_grad():
    frame = sample_frame.unsqueeze(0).to(device)
    mel = sample_mel.unsqueeze(0).to(device)
    out = model(frame, mel)
    probs = F.softmax(out, dim=1)
    conf, pred = torch.max(probs, dim=1)
    prevent_action(dataset.categories[pred.item()], conf.item())
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import models, transforms
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

# -----------------------------
# Dataset Loader
# -----------------------------
class HarmfulDataset(Dataset):
    def __init__(self, frames_dir, mels_dir, transform=None):
        self.frames_dir = frames_dir
        self.mels_dir = mels_dir
        self.samples = []
        self.categories = ['Adult Content', 'Harmful Content', 'Safe', 'Suicide']
        self.label_map = {c: i for i, c in enumerate(self.categories)}
        self.transform = transform

        for label in self.categories:
            frame_path = os.path.join(frames_dir, label)
            mel_path = os.path.join(mels_dir, label)
            frame_files = os.listdir(frame_path)
            mel_files = os.listdir(mel_path)
            for f in frame_files:
                if f in mel_files:
                    self.samples.append((os.path.join(frame_path, f),
                                         os.path.join(mel_path, f),
                                         label))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        frame_path, mel_path, label_name = self.samples[idx]
        frames = np.load(frame_path).astype(np.float32) / 255.0
        mel = np.load(mel_path).astype(np.float32)

        if self.transform:
            frames = self.transform(torch.tensor(frames))

        label = self.label_map[label_name]
        return torch.tensor(frames), torch.tensor(mel), torch.tensor(label)


# -----------------------------
# Model: ResNet + BiLSTM + Attention
# -----------------------------
class AttentionBlock(nn.Module):
    def __init__(self, feature_dim):
        super(AttentionBlock, self).__init__()
        self.attention = nn.Sequential(
            nn.Linear(feature_dim, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )

    def forward(self, x):
        attn_weights = torch.softmax(self.attention(x), dim=1)
        weighted_output = torch.sum(attn_weights * x, dim=1)
        return weighted_output, attn_weights


class HarmfulVideoClassifier(nn.Module):
    def __init__(self, num_classes=4):
        super(HarmfulVideoClassifier, self).__init__()
        self.resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
        self.resnet.fc = nn.Identity()  # Remove final FC layer

        self.lstm = nn.LSTM(2048, 256, num_layers=2, bidirectional=True, batch_first=True, dropout=0.3)
        self.attention = AttentionBlock(512)
        self.fc = nn.Linear(512, num_classes)
        self.dropout = nn.Dropout(0.4)

    def forward(self, frames):
        # Assuming input shape: (batch, seq_len, C, H, W)
        B, T, C, H, W = frames.shape
        frames = frames.view(B * T, C, H, W)
        with torch.no_grad():
            features = self.resnet(frames)
        features = features.view(B, T, -1)
        lstm_out, _ = self.lstm(features)
        attn_out, _ = self.attention(lstm_out)
        out = self.dropout(attn_out)
        return self.fc(out)


# -----------------------------
# Training and Evaluation
# -----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
])

frames_dir = r"E:\archive\Dataset\processed_data\frames"
mels_dir = r"E:\archive\Dataset\processed_data\mels"

dataset = HarmfulDataset(frames_dir, mels_dir, transform=transform)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)

model = HarmfulVideoClassifier(num_classes=4).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.8)

# -----------------------------
# Training Loop
# -----------------------------
num_epochs = 20
best_val_acc = 0
train_losses, val_losses, val_accuracies = [], [], []

for epoch in range(num_epochs):
    model.train()
    running_loss = 0
    for frames, mels, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}", ncols=100):
        frames, labels = frames.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(frames)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    model.eval()
    val_loss, correct, total = 0, 0, 0
    with torch.no_grad():
        for frames, mels, labels in val_loader:
            frames, labels = frames.to(device), labels.to(device)
            outputs = model(frames)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_acc = 100 * correct / total
    val_accuracies.append(val_acc)
    train_losses.append(running_loss / len(train_loader))
    val_losses.append(val_loss / len(val_loader))

    print(f"Epoch {epoch+1}/{num_epochs}: TrainLoss={running_loss/len(train_loader):.4f}, "
          f"ValLoss={val_loss/len(val_loader):.4f}, ValAcc={val_acc:.2f}%")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), "best_harmful_model.pth")

    scheduler.step()

# -----------------------------
# Visualization
# -----------------------------
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.legend()
plt.title("Loss over Epochs")

plt.subplot(1, 2, 2)
plt.plot(val_accuracies, label='Validation Accuracy')
plt.legend()
plt.title("Validation Accuracy over Epochs")
plt.show()

print(f"✅ Best Validation Accuracy: {best_val_acc:.2f}%")
```
